```python
import re
#read the keywords file
with open('Nature_astronomy_keywords2.txt', 'r') as f:
    lin=f.readlines()
```

```python
ascii_list=re.findall(r"u'(.*?)'", lin[0])
```

```python
compact_keywords_list=[]
for strr in ascii_list:
    compact_keywords_list.append(''.join(x for x in strr.title() if not x.isspace()))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# lower max_font_size
wordcloud = WordCloud().generate(''.join(compact_keywords_list))
#wordcloud = WordCloud().generate(''.join(ascii_list))
plt.figure(figsize=(40,40))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
